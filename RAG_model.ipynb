```python
from langchain_core.documents import Document

# --- DOCUMENT 1: ARTIFICIAL INTELLIGENCE ---
ai_content = """
Artificial Intelligence: Foundations, Evolution, and Modern Paradigms

Artificial Intelligence (AI) represents one of the most transformative endeavors in modern computer science, aiming to create systems capable of performing tasks that typically require human intelligence. These tasks include visual perception, speech recognition, decision-making, and translation between languages. The historical trajectory of AI is often characterized by cycles of intense optimism followed by periods of disappointment and reduced funding, commonly known as "AI winters." The early foundations were laid in the 1950s by pioneers like Alan Turing, who proposed the famous "Turing Test" to evaluate a machine's ability to exhibit intelligent behavior, and John McCarthy, who officially coined the term "Artificial Intelligence" at the Dartmouth Conference in 1956.

For several decades, the field was dominated by symbolic AI and expert systems. These classical systems relied on hardcoded rules, formal logic, and vast explicit knowledge bases curated by human experts to solve specific problems. While highly successful in deterministic environments like chess or basic diagnostic systems, symbolic AI struggled immensely with the inherent ambiguity, noise, and scaling challenges of real-world data.

The profound paradigm shift that defines contemporary AI occurred with the rise of Machine Learning (ML) and, subsequently, Deep Learning (DL). Instead of explicitly programming rules, engineers shifted toward designing algorithms that allow computers to discover patterns and learn rules directly from data. Machine learning is broadly categorized into supervised learning (training on labeled datasets), unsupervised learning (finding hidden structures in unlabeled data), and reinforcement learning (learning optimal behaviors through trial-and-error interactions with an environment to maximize a reward).

The modern era of AI is powered by Deep Learning, a subfield utilizing artificial neural networks inspired by the biological structure of the human brain. Deep neural networks consist of an input layer, an output layer, and multiple hidden layers that automatically extract increasingly abstract hierarchies of features from raw data. The availability of massive datasets, coupled with the computational parallelism offered by modern Graphics Processing Units (GPUs) and Tensor Processing Units (TPUs), has enabled the training of models with billions—and now trillions—of parameters. Today, AI systems drive autonomous vehicles, power sophisticated healthcare diagnostic tools, optimize complex global supply chains, and act as foundational components across virtually every major industry, shifting AI from an academic pursuit into a critical utility of the modern digital economy.
"""

document_1 = Document(
    page_content=ai_content.strip(),
    metadata={"source": "resource_library", "topic": "Artificial Intelligence", "length": "long"}
)


# --- DOCUMENT 2: NATURAL LANGUAGE PROCESSING ---
nlp_content = """
Natural Language Processing: Bridging Human Communication and Machine Comprehension

Natural Language Processing (NLP) is a specialized interdisciplinary subfield of artificial intelligence, linguistics, and computer science focused on the interactions between computers and human languages. The ultimate objective of NLP is to enable computers to process, analyze, understand, and generate human language in a way that is both meaningful and contextually valuable. Human language is exceptionally complex, filled with idioms, metaphors, structural ambiguities, syntactic variations, and evolving cultural contexts, making text processing one of the most challenging domains in computer science.

Historically, NLP relied heavily on rule-based computational linguistics, where human experts wrote elaborate grammars and morphological rules to parse text. This evolved into statistical NLP in the 1990s and 2000s, where probabilistic models like Hidden Markov Models (HMMs) and Support Vector Machines (SVMs) were used for tasks like Part-of-Speech (POS) tagging, Named Entity Recognition (NER), and text classification. A major breakthrough occurred with the introduction of word embeddings, such as Word2Vec and GloVe, which mapped words into dense continuous vector spaces where geometric distance reflected semantic similarity.

The true revolution in NLP arrived with the advent of the Transformer architecture, introduced by Vaswani et al. in the seminal 2017 paper "Attention Is All You Need." The Transformer abandoned recurrent neural network (RNN) and long short-term memory (LSTM) designs in favor of the "Self-Attention" mechanism. This allowed models to compute relationships between all words in a sentence simultaneously, capturing long-range dependencies and context far more effectively while allowing massive parallelization during training.

The deployment of Transformers led directly to the creation of Large Language Models (LLMs) like BERT, GPT, and subsequent generations. These models are pre-trained on colossal corpora of internet text using self-supervised objectives (such as predicting missing words or the next token in a sequence). Through this pre-training, models develop an intricate understanding of grammar, facts, and reasoning capabilities. These models are then fine-tuned or aligned using techniques like Reinforcement Learning from Human Feedback (RLHF) to perform complex downstream tasks, including abstractive summarization, conversational dialogue, sentiment analysis, and seamless machine translation.
"""

document_2 = Document(
    page_content=nlp_content.strip(),
    metadata={"source": "resource_library", "topic": "Natural Language Processing", "length": "long"}
)


# --- DOCUMENT 3: RETRIEVAL-AUGMENTED GENERATION ---
rag_content = """
Retrieval-Augmented Generation: Architecting Knowledge-Grounded Language Systems

Retrieval-Augmented Generation (RAG) is an advanced architectural pattern designed to optimize and ground the outputs of Large Language Models (LLMs) by integrating an external, dynamic knowledge retrieval system into the generation process. While LLMs possess remarkable linguistic fluency and expansive general knowledge acquired during training, they suffer from significant operational limitations. These include "hallucinations" (generating plausible-sounding but factually incorrect information), an inability to access private or proprietary enterprise data, and an absolute knowledge cutoff date determined by when the model was last trained. RAG elegantly solves these vulnerabilities without the exorbitant time and computational expense required to continuously fine-tune the model.

The standard operational workflow of a RAG pipeline can be broken down into three core phases: Ingestion, Retrieval, and Generation.

During the Ingestion phase, raw unstructured documents (such as PDFs, manuals, or web pages) are collected and processed. Because LLMs have strict context window limitations, these large documents are broken down into smaller, logically coherent segments called "chunks." These chunks are passed through an embedding model (such as a Hugging Face or OpenAI embedding model) that converts the textual data into high-dimensional numerical vectors. These vectors capture the semantic essence of the text and are indexed and stored inside a specialized Vector Database (such as DataStax Astra DB, Pinecone, or Chroma).

In the Retrieval phase, when a user submits a query to the system, the query is immediately converted into a vector using the exact same embedding model. The vector database performs a mathematical similarity search (typically utilizing Cosine Similarity or Euclidean Distance) to compare the user's query vector against the stored document chunk vectors. The database quickly fetches the top 'K' most semantically relevant chunks that contain the specific information needed to answer the query.

Finally, in the Generation phase, the retrieved document chunks are bundled alongside the user's original query into a carefully structured prompt template. This hybrid prompt provides explicit context to the LLM, effectively instructing it: "Answer the user's question using only the verified facts provided in the following reference blocks." The LLM reads the context, extracts the accurate answers, and synthesizes a naturally phrased response. By grounding the language model's output directly in retrieved source materials, RAG drastically minimizes hallucinations, establishes auditability by allowing citations back to the source text, and ensures the application delivers real-time, context-aware utility.
"""

document_3 = Document(
    page_content=rag_content.strip(),
    metadata={"source": "resource_library", "topic": "Retrieval-Augmented Generation", "length": "long"}
)

# --- CONTEXT LIST CONSOLIDATION ---
# All long-form documents grouped together for your vector store or text splitter
documents = [document_1, document_2, document_3]

# Verification print statement to ensure they loaded properly in Colab
print(f"Successfully loaded {len(documents)} extensive resource documents into Colab context.")
print(f"Topics available: {[doc.metadata['topic'] for doc in documents]}")
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
# 👇 Notice the updated imports here 👇
from langchain_classic.chains import create_history_aware_retriever, create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

# --- PHASE 1: HISTORY-AWARE RETRIEVER ---
contextualize_q_system_prompt = (
    "Given a chat history and the latest user question "
    "which might reference context in the chat history, "
    "formulate a standalone question which can be understood "
    "without the chat history. Do NOT answer the question, "
    "just reformulate it if needed and otherwise return it as is."
)

contextualize_q_prompt = ChatPromptTemplate.from_messages([
    ("system", contextualize_q_system_prompt),
    MessagesPlaceholder("chat_history"),
    ("human", "{input}"),
])

# Create the retriever that dynamically rewrites queries based on history
history_aware_retriever = create_history_aware_retriever(
    llm, retriever, contextualize_q_prompt
)

# --- PHASE 2: QA CHAIN (DOCUMENT STUFFING) ---
qa_system_prompt = (
    "You are an expert technical assistant. Answer the user's question "
    "using exclusively the provided retrieved context below. If you do "
    "not know the answer or if it's not in the context, say that you don't know.\n\n"
    "Context:\n{context}"
)

qa_prompt = ChatPromptTemplate.from_messages([
    ("system", qa_system_prompt),
    MessagesPlaceholder("chat_history"),
    ("human", "{input}"),
])

# Chain to handle document processing
question_answer_chain = create_stuff_documents_chain(llm, qa_prompt)

# --- PHASE 3: COMBINE INTO A RETRIEVAL CHAIN ---
rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)

# --- PHASE 4: STATE MANAGEMENT (CHAT HISTORY) ---
store = {}

def get_session_history(session_id: str) -> InMemoryChatMessageHistory:
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

# Final runnable executable that wraps the chain with session management
conversational_rag_chain = RunnableWithMessageHistory(
    rag_chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history",
    output_messages_key="answer",
)

print("✅ Conversational RAG Chain built successfully!")
```

```python
# Define a unique session ID for the conversation
config = {"configurable": {"session_id": "user_session_01"}}

# --- Turn 1 ---
print("--- User: What is Retrieval-Augmented Generation? ---")
response1 = conversational_rag_chain.invoke(
    {"input": "What is Retrieval-Augmented Generation?"},
    config=config
)
print(f"GPT: {response1['answer']}\n")

# --- Turn 2 (Testing Context History) ---
print("--- User: What are its three core phases? ---")
# The chain automatically rewrites "its" to "Retrieval-Augmented Generation's" using history
response2 = conversational_rag_chain.invoke(
    {"input": "What are its three core phases?"},
    config=config
)
print(f"GPT: {response2['answer']}\n")
```

```python
# 1. Define a unique session ID for the conversation
config = {"configurable": {"session_id": "user_session_01"}}

print("🤖 Chatbot initialized! Type 'exit' or 'quit' to end the conversation.\n")

# 2. Continuous conversation loop
while True:
    # Take dynamic input from the user in Colab
    user_input = input("User: ")

    # Check if the user wants to break out of the loop
    if user_input.lower() in ['exit', 'quit']:
        print("Bot: Goodbye!")
        break

    # Skip empty inputs
    if not user_input.strip():
        continue

    try:
        # Pass the input and session configuration to the chain
        response = conversational_rag_chain.invoke(
            {"input": user_input},
            config=config
        )

        # Print the response from Gemma 3
        print(f"Gemma 3: {response['answer']}\n")
        print("-" * 50) # Visual separator for turns

    except Exception as e:
        print(f"An error occurred: {e}\n")
```
